```python
#aither : returns  asynch iterator from an asyng iterable

import asyncio

async def async_countdown():
    for i in [3, 2, 1]:
        await asyncio.sleep(0.5)  # Simulate network/IO delay
        yield f"Count: {i}"
async def main():
    iterator = aiter(async_countdown())
    print(await anext(iterator))  # Output: Count: 3
    print(await anext(iterator))  # Output: Count: 2
    print(await anext(iterator))  # Output: Count: 1
    
    # 4. The next call will raise a StopAsyncIteration exception
    try:
        await anext(iterator)
    except StopAsyncIteration:
        print("Stream complete!")

asyncio.run(main())
```

```python
# all :returns tru if all elements of iterable are true

def all(iterable):
    for element in iterable:
        if not element:
            return False
    return True
        
```

```python
# any return True is any element of iterable is true.

def any(iterable):
    for element in iterable:
        if element:
            return True
    return False
```

```python
"""
@classmethod
Transform a method into a class method.

A class method receives the class as an implicit first argument,
just like an instance method receives the instance. 
To declare a class method, use this idiom:
"""

from datetime import date

class User:
    # Class-level variable shared by all instances
    total_users = 0

    def __init__(self, name, age):
        self.name = name          # Instance variable
        self.age = age            # Instance variable
        User.total_users += 1

    # 1. Class method serving as an alternative constructor
    @classmethod
    def from_birth_year(cls, name, birth_year):
        calculated_age = date.today().year - birth_year
        # cls(...) is equivalent to calling User(name, calculated_age)
        return cls(name, calculated_age)

    # 2. Class method accessing class-level state
    @classmethod
    def get_total_users(cls):
        return f"Total active users: {cls.total_users}"

# Using the standard constructor
user1 = User("Alice", 30)

# Using the alternative constructor class method
user2 = User.from_birth_year("Bob", 1095)

# Calling a class method to see the global state
print(User.get_total_users())  # Output: Total active users: 2
```

```python
#Compile: Compile the source into a code or AST object.
code_str = """
def greet(name):
    return f"Hello, {name}!"

result = greet("Alice")
print(result)
"""

# 2. Compile into an executable code object
compiled_code = compile(code_str, filename="<string>", mode="exec")

# 3. Execute using exec()
exec(compiled_code)
```

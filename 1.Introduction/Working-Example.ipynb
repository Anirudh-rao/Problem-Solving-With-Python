2. Perimeter and Area Calculator

Write a program that accepts the width and length of a rectangular space
from the user and then calculates both the perimeter and area of this
space. For example, the output of the program could look like this:

````python
width = int(input("Enter the width for the rectangle"))
length = int(input("Enter the length of the rectangle"))

Area = length * width
Perimeter = 2*(length + width)

print(f"The area of the rectangle :{Area}")
print(f"The Perimeter of the rectangle :{Perimeter}")
````
